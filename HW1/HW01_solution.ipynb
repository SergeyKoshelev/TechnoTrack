## Постановка задачи
Пусть имеется набор из N экспериментов. Каждому эксперименту соответствует набор из k переменных (**признаков**), одинаковых по своей сути для всех экспериментов, и также результат этого эксперимента (**целевая переменная**).  
При постановке задачи вида **линейной регрессии** мы пытаемся аппроксимировать целевую переменную **y** линейным образом по признакам **X**, то есть найти такой набор параметров $\theta$ и b, чтобы 
$$
y = X\theta + b,
$$
или в более привычном виде с размерностью переменных и параметров = 1:
$$
y = k\cdot x + b
$$
Очевидно, что ровно такие параметры чаще всего подобрать невозможно, поэтому целесообразно ввести некоторую метрику погрешности аппроксимизации - **фунцию ошибки**, которая характеризует величину отклонения линейной модели от реальных значений. В своем решении я выбрал **среднеквадратичное отклонение (MSE)**.

После обучения модели, следует удостовериться в её правильности и как-то оценить полученный результат. Для этого используется **метрика качестве модели**, которая показывает, насколько подобранные параметры модели соответствуют теоретически рассчитанной зависимости. В своем решении я выбрал среднюю абсолютную процентную ошибку (**MAPE**), по которой сразу определяется относительное отклонение.

## Подготовка
Чтобы применять градиентный спуск только к одному вектору $\theta$, добавим нулевым столбцом столбец единиц. Теперь модель описывается уравнением $y=X\theta$, то есть мы избавились от дополнительного вектора b.

Чтобы после обучения модели проверить её на совершенно новой для неё выборке, проверить её переобученность, разделим всю нашу выбору на тренировочную и тестовую в соотношении 80:20 с помощью готовой функции train_test_split из библиотеки sklearn.model_selection.

```python
def add_bias_column(matrix) :
    rows = matrix.shape[0]
    columns = matrix.shape[1]
    first_column = np.full(shape = (rows,1), fill_value = 1)
    new_matrix = np.hstack((first_column, matrix))
    return new_matrix
```

```python
X_bias = add_bias_column(X1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_bias, y, random_state=42, test_size=0.2)
```

## Классы ошибки и модели, цикл обучения
Используются 2 основных класса - класс **loss**, соответствующий функции ошибки (**MSE**) и класс **linear_regression**, соотвествующий модели линейной регрессии (в нем хранятся параметры $\theta$)

Цикл обучения **train_loop** принимает: матрицу признаков **X**, целевую переменную **y**, объект модели **model**, объект функции ошибки **loss_fn**, а также количество итераций **epochs** (по умолчанию 1000) и параметр шага линейной регресии **alpha** (по умолчанию 1).  
В процессе обучения оптимизируются параметры $\theta$ модели **loss_fn** методом градиентного спуска. При первой итерации все параметры инициализируются нулем. Так как направление антиградиента соответствует направлению наискорейшего уменьшения функции, то на каждой последующей итерации $\theta$ уменьшается на вектор градиента $\nabla_{\theta}{\mathscr{L}}$ фунции потерь, умноженного на шаг alpha. Таким образом, с каждой новой итерацией делается шаг к уменьшению ошибки.

```python
class Differentiable:
    def __init__(self):
        pass
    
    def forward(self, **kwargs):
        raise NotImplementedError()
    
    def backward(self, **kwargs):
        raise NotImplementedError()
```

```python
class loss(Differentiable):
    def __init__(self):
        super(loss, self).__init__()
    
    def forward(self, y_pred, y_true):
        # Этот метод реализует вычисление значения функции потерь (скаляр)
        
        dim = y_true.shape[0]
        difference = np.transpose(y_true - y_pred)
        loss_value = np.inner(difference, difference) / dim #квадрат c нормировкой на размерность
        return loss_value[0][0]
    
    
    def backward(self, y_pred, y_true):
        # Этот метод реализует вычисление градиента функции потерь по аргументу y_pred (вектор с размерностью как у y_pred)
       
        partial_grad = 2 * (y_pred-y_true)
        return partial_grad
```

```python
class linear_regression(Differentiable):
    def __init__(self):
        super(linear_regression, self).__init__()
        self.theta = None
    
    def __call__(self, X):
        # этот метод предназначен для вычисления значения целевой переменной
        return self.forward(X)
    
    def backward(self, X):
        # Этот метод реализует вычисление компоненты градиента функции потерь
        
        partial_grad = np.copy(np.transpose(X))
        return partial_grad
    
    def forward(self, X):
        # этот метод предназначен для применения модели к данным
        assert X.ndim == 2, "X should be 2-dimensional: (N of objects, n of features)"
        
        if (self.theta is None):
            # Если вектор параметров еще не инициализирован, его следует инициализировать
            
            dim = X.shape[1]
            self.theta = np.zeros((dim, 1)) #столбец высоты размером количество столбцов в X
        
        #применение модели к данным, учтен первый единичный столбец у матрицы X
       
        y_pred = np.dot(X, self.theta)
        return y_pred
```

```python
def train_loop(X, y, model, loss_fn, epochs=1000, alpha = 1):
    loss_history = []
    pbar = tqdm(total=epochs)
    for epoch in range(epochs):
        y_pred = model(X)
        loss_value = loss_fn.forward(y_pred, y)
        
        pbar.update(1)
        pbar.set_postfix({'loss': loss_value})
        if (np.isnan(loss_value) or np.isinf(loss_value)):  #при слишком больших значениях alpha градиентный
            print('Bad loss value: {}'.format(loss_value))  #cпуск уходит на бесконечность
            pbar.close()
            return None
        
        grad = np.dot(model.backward(X), loss_fn.backward(y_pred, y))  #получаем значение градиента по правилу сложной функции
        model.theta -= alpha * grad
        loss_history.append(loss_value)

    pbar.close()
    return loss_history
```

## Обучение
Для начала следует найти хотя бы примерное оптимальное значение шага градиентного спуска - параметр alpha. Для этого в цикле обучим модель при разных alpha на 10_000 итерациях, выберем то значение, при котором значение потери оказалось минимальным.

Оказалось, что такое alpha = 1e-9. Обучим модель на большем количестве итераций, например, 100_000.

```python
alphas = [10 ** (-i) for i in range(8, 12)] #найдем примерное значение шага градиентного спуска
for alpha in alphas : 
    obj_fn = loss()
    lr_model = linear_regression()
    loss_history = train_loop(X_train, y_train, lr_model, obj_fn, epochs=10_000, alpha = alpha)
    if loss_history is not None : 
        last_loss = loss_history[-1]
        print("alpha: {}\nloss: {}\n".format(alpha, last_loss))
```

```python
obj_fn = loss()
lr_model = linear_regression()
epochs = 100_000
loss_history = train_loop(X_train, y_train, lr_model, obj_fn, epochs=epochs, alpha = 1e-09)
train_loss = loss_history[-1]
```

## Проверка модели
Для начала исследуем графики величины ошибки от итерации в цикле. 

1-ый график нам мало о чем говорит, так как на первых 200 итерациях наблюдалось очень резкое уменьшение.

По 2-ому графику уже видна экспоненциальная зависимость, которая и соответствует нашим предположениям.

Наконец 3-ий график на последних 5000 итерациях показывает, что зависимость стала около линейной. В целом, на таком количестве итераций нам уже удалось добиться хорошего успеха, поэтому больше обучать модель мы не будем.

## Тестовая выборка
Применим получившуюся модель к оставшейся тестовой выборке. Также выведем среднее значение целевой переменной, чтобы примерно оценить величину ошибки.

В качестве метрики используем **MAPE** и выведем среднее относительное отклонение.

```python
def MAPE (y_test, y_pred) :
    difference_abs = np.abs(y_test - y_pred)
    errors = difference_abs / y_test
    return errors.mean()

prediction = lr_model(X_test)
test_loss = obj_fn.forward(y_test, prediction)
print('Y_mean: {:.4f}'.format(y.mean()))
print('Loss in train dataset: {:.4f}'.format(train_loss))
print('Loss on test dataset: {:.4f}'.format(test_loss))
print('MAPE: {:.1f}%'.format(MAPE(y_test, prediction) * 100))
```

Возможно, стоило закончить обучении на гораздо меньшем числе **epochs**. Метрика **MAPE** показало относительную погрешность предсказания **y** около 6.7%, что является отличным результатом для такой простой модели. Однако, для серьезного увеличения точности понадобится другая, более подходящая модель.
